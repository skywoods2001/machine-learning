```python
import numpy as np
import pandas as pd


train_data = pd.read_csv("train.csv")
bid_history = pd.read_csv("bids.csv")
```

```python
human_ids = train_data[train_data['outcome'] == 0]['bidder_id'].tolist()
robot_ids = train_data[train_data['outcome'] == 1]['bidder_id'].tolist()
print "number of bidders = {}".format(len(train_data))
print "number of humans = {}".format(len(human_ids))
print "number of robots = {}".format(len(robot_ids))
```

```python
num_bids = bid_history['bidder_id'].value_counts().to_frame()
num_bids.columns = ['num_bids']
```

```python
df = train_data.loc[:,['bidder_id','outcome']].join(num_bids, on = 'bidder_id', lsuffix='_l')
```

```python
df = df[~df.num_bids.isnull()]
print df.sort_values(['num_bids'], ascending=False)
```

```python
#df[df['outcome'] == 1].num_bids.plot(kind = 'hist', range= [0,500], bins = 100)
#plt.show()
print "Number of bids made by robots"
print df[df['outcome'] == 1].num_bids.describe()
```

```python
#df[df['outcome'] == 0].num_bids.plot(kind = 'hist', range= [0,500], bins = 100)
#plt.show()
print "Number of bids made by humans"
print df[df['outcome'] == 0].num_bids.describe()
```

```python
bid_history = bid_history.sort_values(by=['bidder_id', 'time'])
bid_history
```

```python
bid_history['time_diff'] = bid_history.groupby('bidder_id')['time'].diff()
```

```python
bid_history
```

```python
min_time_diff = bid_history.groupby('bidder_id')['time_diff'].min()
min_time_diff = pd.DataFrame({'bidder_id': min_time_diff.index, 'min_time_diff': min_time_diff.values})
min_time_diff
```

```python
df = pd.merge(df,min_time_diff, on = 'bidder_id', how = 'left')
```

```python
mean_time_diff = bid_history.groupby('bidder_id')['time_diff'].mean()
mean_time_diff = pd.DataFrame({'bidder_id': mean_time_diff.index, 'mean_time_diff': mean_time_diff.values})
mean_time_diff
```

```python
df = pd.merge(df,mean_time_diff, on = 'bidder_id', how = 'left')
```

```python
median_time_diff = bid_history.groupby('bidder_id')['time_diff'].median()
median_time_diff = pd.DataFrame({'bidder_id': median_time_diff.index, 'median_time_diff': median_time_diff.values})
median_time_diff
```

```python
df = pd.merge(df,median_time_diff, on = 'bidder_id', how = 'left')
```

```python
var_time_diff = bid_history.groupby('bidder_id')['time_diff'].var()
var_time_diff = pd.DataFrame({'bidder_id': var_time_diff.index, 'var_time_diff': var_time_diff.values})
var_time_diff
```

```python
df = pd.merge(df,var_time_diff, on = 'bidder_id', how = 'left')
```

```python
df
```

```python
num_auction = bid_history.groupby('bidder_id')['auction'].unique().apply(len)
num_auction = pd.DataFrame({'bidder_id': num_auction.index, 'num_auction': num_auction.values})
num_auction
```

```python
df = pd.merge(df,num_auction, on = 'bidder_id', how = 'left')
```

```python
num_url = bid_history.groupby('bidder_id')['url'].unique().apply(len)
num_url = pd.DataFrame({'bidder_id': num_url.index, 'num_url': num_url.values})
num_url
```

```python
df = pd.merge(df,num_url, on = 'bidder_id', how = 'left')
```

```python
df['num_bids_per_auction'] = df['num_bids'] / df['num_auction']
```

```python
df['num_bids_per_url'] = df['num_bids'] / df['num_url']
```

```python
num_ip = bid_history.groupby('bidder_id')['ip'].unique().apply(len)
num_ip = pd.DataFrame({'bidder_id': num_ip.index, 'num_ip': num_ip.values})
df = pd.merge(df,num_ip, on = 'bidder_id', how = 'left')
```

```python
num_bids_auction = bid_history.groupby(['bidder_id','auction']).apply(len)
```

```python
num_bids_auction_median = num_bids_auction.median(level=[0])
num_bids_auction_median = pd.DataFrame({'bidder_id': num_bids_auction_median.index, 'num_bids_auction_median': num_bids_auction_median.values})
df = pd.merge(df,num_bids_auction_median, on = 'bidder_id', how = 'left')
```

```python
winner_id = bid_history.loc[bid_history.groupby('auction')['time'].idxmax()].bidder_id
win_counts = pd.DataFrame({'bidder_id': winner_id.value_counts().index, 'win_counts': winner_id.value_counts().values})
win_counts
```

```python
df = pd.merge(df,win_counts, on = 'bidder_id', how = 'left')
```

```python
df.win_counts = df.win_counts.fillna(0)
df['win_prob'] = df['win_counts'] / df['num_auction']
```
